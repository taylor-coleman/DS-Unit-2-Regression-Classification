```
import numpy as np
import pandas as pd

# Read New York City apartment rental listing data
df = pd.read_csv('../data/renthop-nyc.csv')
assert df.shape == (49352, 34)

# Remove the most extreme 1% prices,
# the most extreme .1% latitudes, &
# the most extreme .1% longitudes
df = df[(df['price'] >= np.percentile(df['price'], 0.5)) & 
        (df['price'] <= np.percentile(df['price'], 99.5)) & 
        (df['latitude'] >= np.percentile(df['latitude'], 0.05)) & 
        (df['latitude'] < np.percentile(df['latitude'], 99.95)) &
        (df['longitude'] >= np.percentile(df['longitude'], 0.05)) & 
        (df['longitude'] <= np.percentile(df['longitude'], 99.95))]
```

```
# To takeout outliers for baths and beds
df = df.query('bedrooms <= 5 and bathrooms <= 5')
print(df.shape)
df.head()
```

```
df['created'] = pd.to_datetime(df['created'], infer_datetime_format=True)
```

```
boolfeatures = df.columns.tolist()
del boolfeatures[:10]
df['amenities'] = df[boolfeatures].sum(axis=1)
df.head()
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# split data
train = df[(df['created'].dt.month == 4) | (df['created'].dt.month == 5)]
test = df[df['created'].dt.month == 6]

# Identifying features and targets
features = ['bathrooms','bedrooms','amenities']
target = 'price'

# X and y for train and test.
X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]

# fit model
model = LinearRegression()
model.fit(X_train,y_train)
```
